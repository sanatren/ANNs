```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#Text example
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Machine learning is a fascinating field of study.",
    "Word embeddings help in capturing semantic meanings of words.",
    "I love exploring new ideas and concepts.",
    "The weather today is quite pleasant and sunny."
]
```

```python
vocab_size = 10000 #len(sentences)
```

```python
ohe_repr = [one_hot(words, vocab_size) for words in sentences]
ohe_repr
```

```python
#importing libraries for word embeddings
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import  pad_sequences
from tensorflow.keras.layers import Embedding
import numpy as np
```

```python
#finding maximum length of a sentence can be.
sent_len = max(len(sentences) for sentences in ohe_repr)
sent_len
```

```python
#adding 0 to make all the arrays in same dimension
padded_form = pad_sequences(ohe_repr,padding='pre', maxlen=sent_len)
print(padded_form)
```

```python
#adding embedding to each word and train accordingly
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=10, input_length=sent_len))
```

```python
#train
model.compile('adam','mse')
```

```python
#prediction
model.predict(padded_form)
```

```python
from sentence_transformers import SentenceTransformer

#Example sentences
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Machine learning is a fascinating field of study.",
    "Word embeddings help in capturing semantic meanings of words.",
    "I love exploring new ideas and concepts.",
    "The weather today is quite pleasant and sunny."
]

#Load a pre-trained model for embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')

#Generate embeddings
embeddings = model.encode(sentences)

# Display the embeddings
for i, sentence in enumerate(sentences):
    print(f"Sentence: {sentence}")
    print(f"Embedding: {embeddings[i][:5]}...")  #Display first 5 dimensions for brevity
    print()
```
